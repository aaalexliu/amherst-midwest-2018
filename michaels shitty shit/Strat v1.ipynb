```python
from portfolio import PortfolioGenerator
import pandas as pd
import numpy as np
import statsmodels.formula.api as sm
import math
```

```python
class SampleStrategy(PortfolioGenerator):

    def __init__(self):
        pass

    def build_signal(self, stock_features):
        return self.momentum(stock_features)

    def momentum(self, stock_features):
        return stock_features.groupby(['ticker'])['returns'].mean()
```

```python
# Test out performance by running 'python sample_strategy.py'
if __name__ == "__main__":
    portfolio = SampleStrategy()
    sharpe = portfolio.simulate_portfolio()
    print("*** Strategy Sharpe is {} ***".format(sharpe))
```

```python
factor = pd.read_csv('stock_data/factor_data.csv')
ticker = pd.read_csv('stock_data/ticker_data.csv')
chunks = [factor, ticker]

df = pd.DataFrame.merge(factor, ticker, on = 'timestep', how = "left")
df.set_index('timestep', inplace=True)
df.returns = pd.to_numeric(df.returns)
df = df.assign(book_to_market = 1/df.pb)

#create small/big cap factors

def grouping(data):
    dec=pd.qcut(data['market_cap'],[0,0.5,1],labels=['small_cap', 'big_cap'])
    data['market_cap_ranks'] =  dec
    return data

df = df.groupby('timestep').apply(grouping)

#create book/market ranks

def grouping2(data):
    dec=pd.qcut(data['book_to_market'],[0,0.3,0.7,1],labels=['value', 'neutral', 'growth'])
    data['book_market_ranks'] =  dec
    return data

df = df.groupby('timestep').apply(grouping2)
print(df[1012:1020])
```

```python
#todo: risk free returns?

def fama_french_factors(data):
    market_returns = np.mean(data['returns'])
    small_cap_growth_returns = np.mean(data[(data.market_cap_ranks == 'small_cap') & (data.book_market_ranks == 'growth')]['returns'])
    small_cap_neutral_returns = np.mean(data[(data.market_cap_ranks == 'small_cap') & (data.book_market_ranks == 'neutral')]['returns'])
    small_cap_value_returns = np.mean(data[(data.market_cap_ranks == 'small_cap') & (data.book_market_ranks == 'value')]['returns'])
    big_cap_growth_returns = np.mean(data[(data.market_cap_ranks == 'big_cap') & (data.book_market_ranks == 'growth')]['returns'])
    big_cap_neutral_returns = np.mean(data[(data.market_cap_ranks == 'big_cap') & (data.book_market_ranks == 'neutral')]['returns'])
    big_cap_value_returns = np.mean(data[(data.market_cap_ranks == 'big_cap') & (data.book_market_ranks == 'value')]['returns'])
    data['market_returns'] = market_returns
    data['fama_french_SMB'] = (small_cap_growth_returns+small_cap_neutral_returns+small_cap_value_returns)/3 - (big_cap_growth_returns + big_cap_neutral_returns+ big_cap_value_returns)/3
    data['fama_french_HML'] = (small_cap_value_returns + big_cap_value_returns)/2 - (small_cap_growth_returns+big_cap_growth_returns)/2
    return data

df = df.groupby('timestep').apply(fama_french_factors)

print(df[2020:2025])

'''
market_returns = np.mean(df['returns'])
small_cap_growth_returns = np.mean(df[(df.market_cap_ranks == 'small_cap') & (df.book_market_ranks == 'growth')]['returns'])
small_cap_neutral_returns = np.mean(df[(df.market_cap_ranks == 'small_cap') & (df.book_market_ranks == 'neutral')]['returns'])
small_cap_value_returns = np.mean(df[(df.market_cap_ranks == 'small_cap') & (df.book_market_ranks == 'value')]['returns'])
big_cap_growth_returns = np.mean(df[(df.market_cap_ranks == 'big_cap') & (df.book_market_ranks == 'growth')]['returns'])
big_cap_neutral_returns = np.mean(df[(df.market_cap_ranks == 'big_cap') & (df.book_market_ranks == 'neutral')]['returns'])
big_cap_value_returns = np.mean(df[(df.market_cap_ranks == 'big_cap') & (df.book_market_ranks == 'value')]['returns'])

fama_french_SMB = (small_cap_growth_returns+small_cap_neutral_returns+small_cap_value_returns)/3 - (
    big_cap_growth_returns + big_cap_neutral_returns+ big_cap_value_returns)/3

fama_french_HML = (small_cap_value_returns + big_cap_value_returns)/2 - (small_cap_growth_returns+big_cap_growth_returns)/2
'''
```
