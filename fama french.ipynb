```python
from portfolio import PortfolioGenerator
import pandas as pd
import numpy as np
import math
import itertools
import statsmodels.api as sm
from statsmodels import regression,stats
import scipy
```

```python
class SampleStrategy(PortfolioGenerator):

    def __init__(self):
        pass

    def build_signal(self, stock_features):
        return self.momentum(stock_features)

    def momentum(self, stock_features):
        return stock_features.groupby(['ticker'])['returns'].mean()
```

```python
def read_stock_data():
        '''
        Description:
            Reads in simulated stock data from stock_data.csv
        Returns:
            stock_df (DataFrame): standardized ticker/factor data in pandas df
        Raises:
            AssertionError: ticker_data.csv/factor_data.csv has an invalid format
        '''
        ticker_df = pd.read_csv('stock_data/ticker_data.csv')
        factor_df = pd.read_csv('stock_data/factor_data.csv')
        assert 'timestep' in ticker_df.columns, "ticker_data.csv has an invalid format"
        assert 'ticker' in ticker_df.columns, "ticker_data.csv has an invalid format"
        assert 'returns' in ticker_df.columns, "ticker_data.csv has an invalid format"
        assert 'timestep' in factor_df.columns, "factor_data.csv has an invalid format"
        ticker_df.set_index('timestep', inplace=True)
        factor_df.set_index('timestep', inplace=True)
        stock_df = ticker_df.join(factor_df, how='left')
        return stock_df
```

```python
df = read_stock_data()
```

```python
df = df.assign(book_to_market = 1/df.pb)

#create small/big cap factors

def grouping(data):
    dec=pd.qcut(data['market_cap'],[0,0.5,1],labels=['small_cap','big_cap'])
    data['market_cap_ranks'] =  dec
    return data

df = df.groupby('timestep').apply(grouping)

#create book/market ranks

def grouping2(data):
    dec=pd.qcut(data['book_to_market'],[0,0.3,0.7,1],labels=['growth', 'neutral', 'value'])
    data['book_market_ranks'] =  dec
    return data

df = df.groupby('timestep').apply(grouping2)
```

```python
df = df.rename(columns={'3M_R':'THREEMR'})
df = df.assign(rf_rate = (np.power(1+ df.THREEMR/100,1/90) -1))
```

```python
df.assign(log_returns = np.log(1+df.returns))
```

```python
#todo: risk free returns?

def fama_french_factors(data):
#     market_returns = np.mean(data['returns'])
    small_cap_growth_returns = np.mean(data[(data.market_cap_ranks == 'small_cap') & (data.book_market_ranks == 'growth')]['returns'])
    small_cap_neutral_returns = np.mean(data[(data.market_cap_ranks == 'small_cap') & (data.book_market_ranks == 'neutral')]['returns'])
    small_cap_value_returns = np.mean(data[(data.market_cap_ranks == 'small_cap') & (data.book_market_ranks == 'value')]['returns'])
    big_cap_growth_returns = np.mean(data[(data.market_cap_ranks == 'big_cap') & (data.book_market_ranks == 'growth')]['returns'])
    big_cap_neutral_returns = np.mean(data[(data.market_cap_ranks == 'big_cap') & (data.book_market_ranks == 'neutral')]['returns'])
    big_cap_value_returns = np.mean(data[(data.market_cap_ranks == 'big_cap') & (data.book_market_ranks == 'value')]['returns'])
#     data['market_returns'] = market_returns
    data['fama_french_SMB'] = (small_cap_growth_returns+small_cap_neutral_returns+small_cap_value_returns)/3 - (big_cap_growth_returns + big_cap_neutral_returns+ big_cap_value_returns)/3
    data['fama_french_HML'] = (small_cap_value_returns + big_cap_value_returns)/2 - (small_cap_growth_returns+big_cap_growth_returns)/2
    return data

def proper_fama_french_factors(data):
#     market_returns = np.mean(data['returns'])
    small_cap_returns = np.mean(data[data.market_cap_ranks == 'small_cap']['returns'])
    big_cap_returns = np.mean(data[(data.market_cap_ranks == 'big_cap')]['returns'])
    value_returns = np.mean(data[data.book_market_ranks == 'value']['returns'])
    growth_returns = np.mean(data[data.book_market_ranks == 'growth']['returns'])
    data['EXMARKET'] = data.BIG_IX_pctchg - data.rf_rate
    data['SMB'] = small_cap_returns - big_cap_returns
    data['HML'] = value_returns - growth_returns
    return data

df = df.groupby('timestep').apply(proper_fama_french_factors)
```
